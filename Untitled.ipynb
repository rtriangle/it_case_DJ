```python
def print_correlated_features(X, cor_margin=0.9):
    features = list(X.columns)
    corr_map = X.corr()
    g = nx.Graph()
    for feature in features:
        if sum(corr_map[feature] > cor_margin) > 1:
            indexes = set(corr_map[feature][corr_map[feature] > cor_margin].index)
            indexes.remove(feature)
            print(feature, corr_map[feature][list(indexes)[0]], list(indexes))
            for cor_feature in indexes:
                g.add_edge(feature, cor_feature)
    nx.draw(g)
```

```python
def delete_correlated_features(X, cor_margin=0.9):
    features = list(X.columns)
    bad_features = []
    total_corr_map = X.corr()
    flag = True
    print("Initial features num", len(features))
    while True:
        corr_map = total_corr_map.ix[features, features]
        flag = False
        max_links = 0
        max_linked_features = None
        for feature in features:
            if sum(corr_map[feature] > cor_margin) - 1 > max_links:
                flag = True
                max_links = sum(corr_map[feature] > cor_margin) - 1
                indexes = list(corr_map[feature][corr_map[feature] > cor_margin].index)
                indexes.remove(feature)
                max_linked_features = indexes
        if not flag:
            break
        for feature in max_linked_features:
            features.remove(feature)
        print("deleted", max_links, "features")
    print("Total features", len(features))
    return X[features]
```

```python
from sklearn.feature_selection import mutual_info_classif
def delete_small_mutual_info(X, y, margin=0.001):
    mic = mutual_info_classif(X.fillna(0), y)
    features = np.array(X.columns)[mic > margin]
    return X[features]
```

```python
from sklearn.decomposition import PCA

def apply_PCA(X, n_components=150):
    transformer = PCA(n_components=n_components)
    return transformer.fit_transform(X.fillna(0))
```
